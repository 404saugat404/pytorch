```python
import torch
from torch import nn
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
NUM_CLASSES=8
NUM_FEATURES = 2
RANDOM_SEED = 42

# 1. Create multi-class data
X_blob, y_blob = make_blobs(n_samples=1000,
    n_features=NUM_FEATURES, # X features
    centers=NUM_CLASSES, # y labels
    cluster_std=1.5, # give the clusters a little shake up (try changing this to 1.0, the default)
    random_state=RANDOM_SEED
)

# 2. Turn data into tensors
X_blob = torch.from_numpy(X_blob).type(torch.float)
y_blob = torch.from_numpy(y_blob).type(torch.LongTensor)
print(X_blob[:5], y_blob[:5])

# 3. Split into train and test sets
X_blob_train, X_blob_test, y_blob_train, y_blob_test = train_test_split(X_blob,
    y_blob,
    test_size=0.2,
    random_state=RANDOM_SEED
)

# 4. Plot data
plt.figure(figsize=(10, 7))
plt.scatter(X_blob[:, 0], X_blob[:, 1], c=y_blob, cmap=plt.cm.RdYlBu);
```

```python
#accuracy fn
def accuracy(y_true,y_pred):
  correct=torch.eq(y_true,y_pred).sum().item()
  acc=correct/len(y_pred)*100
  return acc
```

```python
#lets build a model
device="cuda" if torch.cuda.is_available() else "cpu"

class MulClass(nn.Module):
  def __init__(self,input_features,output_features,hidden_units):
    super().__init__()
    self.linear_layer_stack=nn.Sequential(
        nn.Linear(in_features=input_features,out_features=hidden_units),
        nn.ReLU(),
        nn.Linear(in_features=hidden_units,out_features=hidden_units),
        nn.ReLU(),
        nn.Linear(in_features=hidden_units,out_features=output_features)
    )
  def forward(self,x):
    return self.linear_layer_stack(x)
```

```python
m_0=MulClass(input_features=2,output_features=8,hidden_units=8).to(device)
```

```python
loss_fn=nn.CrossEntropyLoss()
optimizer=torch.optim.SGD(m_0.parameters(),lr=0.1)
```

```python
logits=m_0(X_blob_train)
logits[:5]
```

```python
pred=torch.softmax(logits,dim=1)
pred[:5]
```

```python
print(torch.argmax(pred[0]))
print(y_blob_train[0])
```

```python
#training and testing loop
torch.manual_seed(42)
epochs=10000
X_blob_train, y_blob_train = X_blob_train.to(device), y_blob_train.to(device)
X_blob_test, y_blob_test = X_blob_test.to(device), y_blob_test.to(device)

for i in range(epochs):
  m_0.train()

  y_logits=m_0(X_blob_train)

  y_pred=torch.softmax(y_logits,dim=1).argmax(dim=1)

  loss=loss_fn(y_logits,y_blob_train)
  acc=accuracy(y_blob_train,y_pred)

  optimizer.zero_grad()

  loss.backward()

  optimizer.step()


  with torch.inference_mode():
    test_logits=m_0(X_blob_test)
    test_pred=torch.softmax(test_logits,dim=1).argmax(dim=1)

    test_loss=loss_fn(test_logits,y_blob_test)
    test_acc=accuracy(y_blob_test,test_pred)

  if i%1000==0:
    print(f"Epoch: {i} | Loss: {loss:.5f}, Acc: {acc:.2f}% | Test Loss: {test_loss:.5f}, Test Acc: {test_acc:.2f}%")


```
