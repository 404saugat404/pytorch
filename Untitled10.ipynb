```python
import torch
```

```python
#weight and bias
w=0.6
b=6

start=0
end=1
step=0.02
X=torch.arange(start,end,step).unsqueeze(dim=1)
y=w*X+b
```

```python
X[:10],y[:10]
```

```python
length_split=int(0.8*len(X))
X_train,y_train=X[:length_split],y[:length_split]
X_test,y_test=X[length_split:],y[length_split:]
```

```python
#defining the fn for plotting
import matplotlib.pyplot as plt

def prediction(train_data=X_train,
               train_labels=y_train,
               test_data=X_test,
               test_labels=y_test,
               predictions=None):
  plt.scatter(train_data,train_labels,label="Training data",marker="o",c="g")

  plt.scatter(test_data,test_labels,label="Testing data",marker="o",c="b")

  if predictions is not None:
    plt.scatter(test_data,predictions,label="Predictions",marker="o",c="r")

  plt.legend(prop={"size":14})
```

```python
prediction()
```

```python
#building a model
import torch
from torch import nn

class LinReg(nn.Module):
  def __init__(self):
    super().__init__()
    self.weights=nn.Parameter(torch.randn(1,
                                          requires_grad=True,
                                          dtype=torch.float))
    self.bias=nn.Parameter(torch.randn(1,
                                      requires_grad=True,
                                      dtype=torch.float))
  def forward(self,x:torch.Tensor)->torch.Tensor:
    return self.weights*x+self.bias
```

```python
torch.manual_seed(42)
model_0=LinReg()
```

```python
model_0.state_dict()
```

```python
with torch.inference_mode():
  y_preds=model_0(X_test)
```

```python
X_test.size(),y_preds.size()
```

```python
prediction( predictions=y_preds)
```

```python
loss_fn=nn.L1Loss()
optimizer=torch.optim.SGD(params=model_0.parameters(),
                          lr=0.01)
```

```python
#training
torch.manual_seed(42)
epochs=10000

train_loss=[]
test_loss=[]
epoch_count=[]

for i in range(epochs):
  model_0.train()
  y_pred=model_0(X_train)
  loss=loss_fn(y_pred,y_train)
  optimizer.zero_grad()
  loss.backward()
  optimizer.step()

  #testing
  model_0.eval()
  with torch.inference_mode():
    test_pred=model_0(X_test)
    test_loss_value=loss_fn(test_pred,y_test)

  if i%100==0:
    train_loss.append(loss.detach().numpy())
    test_loss.append(test_loss_value.detach().numpy())
    epoch_count.append(i)
    print(f"Epoch: {i} | Loss: {loss} | Test Loss: {test_loss_value}")
```

```python
plt.plot(epoch_count,train_loss,label="Train loss")
plt.plot(epoch_count,test_loss,label="Test loss")
plt.title("Training and testing loss")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.legend()
```
